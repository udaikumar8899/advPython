```python
#1. Parking Lot Management Track vehicle entry/exit, available spots, and
# compute parking fees.
class ParkingLot:
    def __init__(self, total_spots):
        self.total_spots = total_spots
        self.occupied = {}
        self.rate_per_hour = 20

    def vehicle_entry(self, vehicle_no, entry_time):
        if len(self.occupied) < self.total_spots:
            self.occupied[vehicle_no] = entry_time
            print("Vehicle parked successfully")
        else:
            print("Parking Full")

    def vehicle_exit(self, vehicle_no, exit_time):
        if vehicle_no in self.occupied:
            entry_time = self.occupied.pop(vehicle_no)
            hours = exit_time - entry_time
            fee = hours * self.rate_per_hour
            print("Parking Fee:", fee)
        else:
            print("Vehicle not found")

    def available_spots(self):
        print("Available Spots:", self.total_spots - len(self.occupied))
```

```python
#2. To-Do List Manager with Categories Add tasks, set deadlines, assign
#  priority, and filter by project or tag.
class Task:
    def __init__(self, title, deadline, priority, category):
        self.title = title
        self.deadline = deadline
        self.priority = priority
        self.category = category


class TodoManager:
    def __init__(self):
        self.tasks = []

    def add_task(self, title, deadline, priority, category):
        task = Task(title, deadline, priority, category)
        self.tasks.append(task)

    def show_tasks(self):
        for t in self.tasks:
            print(t.title, t.deadline, t.priority, t.category)

    def filter_by_category(self, category):
        for t in self.tasks:
            if t.category == category:
                print(t.title, t.deadline)
```

```python
#3. Billing System for Retail Store Scan products, apply discounts,
#  generate bills, and record transactions.
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price


class BillingSystem:
    def __init__(self):
        self.cart = []

    def add_product(self, name, price):
        product = Product(name, price)
        self.cart.append(product)

    def generate_bill(self, discount=0):
        total = sum(p.price for p in self.cart)
        total = total - (total * discount / 100)

        print("----- BILL -----")
        for p in self.cart:
            print(p.name, p.price)

        print("Total after discount:", total)
```

```python
#4. Movie Ticket Booking System Check showtimes, book tickets, select
# seats, and print confirmation.
class Movie:
    def __init__(self, name, showtime, seats):
        self.name = name
        self.showtime = showtime
        self.available_seats = seats

    def book_ticket(self, seats):
        if seats <= self.available_seats:
            self.available_seats -= seats
            print("Booking confirmed for", seats, "seats")
        else:
            print("Not enough seats available")

    def show_details(self):
        print("Movie:", self.name)
        print("Showtime:", self.showtime)
        print("Available seats:", self.available_seats)
```

```python
#5. College Result Management Teachers enter marks, calculate GPA
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll
        self.marks = []

    def add_marks(self, mark):
        self.marks.append(mark)

    def calculate_gpa(self):
        avg = sum(self.marks) / len(self.marks)

        if avg >= 90:
            gpa = 10
        elif avg >= 80:
            gpa = 9
        elif avg >= 70:
            gpa = 8
        elif avg >= 60:
            gpa = 7
        else:
            gpa = 6

        return gpa

    def display_result(self):
        print("Name:", self.name)
        print("Roll:", self.roll)
        print("Marks:", self.marks)
        print("GPA:", self.calculate_gpa())
```

```python
#6. Password Manager securely store ,add, edit,and retrive user passwords
# using encryption.
class PasswordManager:

    def __init__(self):
        self.passwords = {}

    def encrypt(self, password):
        encrypted = ""
        for ch in password:
            encrypted += chr(ord(ch) + 2)
        return encrypted

    def decrypt(self, password):
        decrypted = ""
        for ch in password:
            decrypted += chr(ord(ch) - 2)
        return decrypted

    def add_password(self, site, password):
        self.passwords[site] = self.encrypt(password)
        print("Password added successfully!")

    def get_password(self, site):
        if site in self.passwords:
            print("Password:", self.decrypt(self.passwords[site]))
        else:
            print("Site not found")

    def edit_password(self, site, new_password):
        if site in self.passwords:
            self.passwords[site] = self.encrypt(new_password)
            print("Password updated successfully!")
        else:
            print("Site not found")

pm = PasswordManager()

while True:
    print("\n1. Add Password")
    print("2. Retrieve Password")
    print("3. Edit Password")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        site = input("Enter site: ")
        pwd = input("Enter password: ")
        pm.add_password(site, pwd)

    elif choice == "2":
        site = input("Enter site: ")
        pm.get_password(site)

    elif choice == "3":
        site = input("Enter site: ")
        pwd = input("Enter new password: ")
        pm.edit_password(site, pwd)

    elif choice == "4":
        break

    else:
        print("Invalid choice")
```

```python
#7. freelancer Marketplace Simulation Register freelancers and clients,
# assign projects, and process payments.
class Person:
    def __init__(self, name):
        self.name = name

class Freelancer(Person):
    def __init__(self, name, skill):
        super().__init__(name)
        self.skill = skill
        self.earnings = 0

    def add_payment(self, amount):
        self.earnings += amount
        print(self.name, "received payment:", amount)

class Client(Person):
    def __init__(self, name):
        super().__init__(name)

class Project:
    def __init__(self, title, budget, client):
        self.title = title
        self.budget = budget
        self.client = client
        self.freelancer = None

    def assign_freelancer(self, freelancer):
        self.freelancer = freelancer
        print("Project assigned to", freelancer.name)

    def complete_project(self):
        if self.freelancer:
            self.freelancer.add_payment(self.budget)
            print("Project completed")
        else:
            print("No freelancer assigned")

freelancers = []
clients = []

fname = input("Enter freelancer name: ")
skill = input("Enter freelancer skill: ")
freelancer = Freelancer(fname, skill)
freelancers.append(freelancer)

cname = input("Enter client name: ")
client = Client(cname)
clients.append(client)

title = input("Enter project title: ")
budget = int(input("Enter project budget: "))
project = Project(title, budget, client)

project.assign_freelancer(freelancer)

project.complete_project()

print("Total earnings:", freelancer.earnings)
```

```python
#8.warehouse automation system : track goods movement ,generate inventory reports and forecaste demand
warehouse automation system

track goods movement , generate inventory reports and forecast demand

class Product:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
        self.history = []


class Warehouse:
    def __init__(self):
        self.products = {}

    def add_product(self, name, qty):
        if name not in self.products:
            self.products[name] = Product(name, qty)
        else:
            self.products[name].quantity += qty

        self.products[name].history.append(("IN", qty))
        print(qty, name, "added to warehouse")


    def remove_product(self, name, qty):
        if name in self.products and self.products[name].quantity >= qty:
            self.products[name].quantity -= qty
            self.products[name].history.append(("OUT", qty))
            print(qty, name, "shipped")
        else:
            print("Not enough stock")


    def inventory_report(self):
        print("\nInventory Report")
        for p in self.products.values():
            print(p.name, ":", p.quantity)


    def movement_history(self, name):
        if name in self.products:
            print("\nMovement history for", name)
            for record in self.products[name].history:
                print(record)


    def forecast_demand(self, name):
        if name in self.products:
            history = self.products[name].history
            out = sum(qty for action, qty in history if action == "OUT")

            if out > 20:
                print(name, "High demand - restock soon")
            else:
                print(name, "Demand normal")

warehouse = Warehouse()

warehouse.add_product("Laptop", 50)
warehouse.add_product("Mouse", 100)

warehouse.remove_product("Laptop", 10)
warehouse.remove_product("Mouse", 25)

warehouse.inventory_report()

warehouse.movement_history("Laptop")

warehouse.forecast_demand("Mouse")
```

```python

#crm (customer relationship manager) store customer info , manage communication logs , and track sales pipelines


class Customer:
    def __init__(self, name, email, phone):
        self.name = name
        self.email = email
        self.phone = phone
        self.logs = []
        self.sales_stage = "Lead"

    def add_log(self, message):
        self.logs.append(message)

    def update_stage(self, stage):
        self.sales_stage = stage


class CRM:
    def __init__(self):
        self.customers = {}

    def add_customer(self, name, email, phone):
        customer = Customer(name, email, phone)
        self.customers[name] = customer
        print("Customer added")

    def add_communication(self, name, message):
        if name in self.customers:
            self.customers[name].add_log(message)
            print("Log added")
        else:
            print("Customer not found")

    def update_sales_stage(self, name, stage):
        if name in self.customers:
            self.customers[name].update_stage(stage)
            print("Stage updated")

    def show_customer(self, name):
        if name in self.customers:
            c = self.customers[name]
            print("Name:", c.name)
            print("Email:", c.email)
            print("Phone:", c.phone)
            print("Sales Stage:", c.sales_stage)
            print("Communication Logs:", c.logs)


crm = CRM()

crm.add_customer("Ravi", "ravi@email.com", "9876543210")
crm.add_communication("Ravi", "Called customer about product")
crm.update_sales_stage("Ravi", "Negotiation")
crm.show_customer("Ravi")
```
