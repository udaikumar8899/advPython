# **transiction system using function**

```python
balance = 0
transactions = []

USERNAME = "udai"
PASSWORD = "143"


def login():
    attempts = 3
    while attempts > 0:
        user = input("Enter username: ")
        pwd = input("Enter password: ")

        if user == USERNAME and pwd == PASSWORD:
            print("\nLogin successful\n")
            return True
        else:
            attempts -= 1
            print("Invalid credentials. Attempts left:", attempts)

    print("\nToo many failed attempts. Account locked.")
    return False


def deposit():
    global balance
    amt = int(input("Enter the amount to deposit: "))
    if amt > 0:
        balance += amt
        transactions.append(f"Deposited: +{amt}")
        print("Amount deposited successfully.")
    else:
        print("Invalid amount.")



def withdraw():
    global balance
    wamt = int(input("Enter amount to withdraw: "))
    if wamt > 0 and balance >= wamt:
        balance -= wamt
        transactions.append(f"Withdrawn: -{wamt}")
        print("Amount withdrawn successfully.")
    else:
        print("Insufficient balance or invalid amount.")

def check_balance():
    print("Your current balance is:", balance)


def transaction_history():
    print("\nTransaction History:")
    if transactions:
        for t in transactions:
            print(t)
    else:
        print("No transactions yet.")


def main_menu():
    while True:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Transaction History")
        print("5. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            deposit()
        elif choice == 2:
            withdraw()
        elif choice == 3:
            check_balance()
        elif choice == 4:
            transaction_history()
        elif choice == 5:
            print("Thank you for using the system.")
            break
        else:
            print("Invalid choice. Please try again.")


if login():
    main_menu()
```
