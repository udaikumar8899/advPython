```python
import time

# Default times
GREEN_TIME = 10
YELLOW_TIME = 3
RED_TIME = 10

signal = input("Enter signal colour (red / yellow / green): ").lower()

if signal == "green":
    print("\n  GREEN LIGHT - GO")
    for i in range(GREEN_TIME, 0, -1):
        print("Time left:", i, "seconds")
        time.sleep(1)

elif signal == "yellow":
    print("\n   YELLOW LIGHT - READY")
    for i in range(YELLOW_TIME, 0, -1):
        print("Time left:", i, "seconds")
        time.sleep(1)

elif signal == "red":
    print("\n  RED LIGHT - STOP")
    for i in range(RED_TIME, 0, -1):
        print("Time left:", i, "seconds")
        time.sleep(1)

else:
    print("  Invalid signal colour entered!")
```
