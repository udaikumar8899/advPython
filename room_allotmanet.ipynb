```python
# Room allotment without using functions

rooms = {}   # dictionary to store room allotment

print("Room Allotment System")
n = int(input("Enter number of students: "))

for i in range(n):
    print("\nStudent", i+1)
    name = input("Enter student name: ")
    year = int(input("Enter year (1 to 4): "))

    if year == 1:
        room = "Room A1"
    elif year == 2:
        room = "Room B1"
    elif year == 3:
        room = "Room C1"
    elif year == 4:
        room = "Room D1"
    else:
        room = "Invalid year"

    rooms[name] = room

print("\n--- Final Room Allotment ---")
for student in rooms:
    print(student, "→", rooms[student])
```
