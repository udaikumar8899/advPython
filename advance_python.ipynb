```python
def run_event_tracker():
    entered_std = set()
    rejected_std = set()

    n = int(input("Enter Entry Attempts: "))

    for i in range(n):
        name = input("Enter students name: ")

        name = name.strip().lower()

        if name in entered_std:
            print(name, "Already entered. Entry Rejected")
            rejected_std.add(name)


        else:
            print(name, "Entry Allowed")
            entered_std.add(name)

    print("-- Event Tracker --")

    print("-- Total Entered Student")
    for name in entered_std:
        print(name)

    print("--Total Rejected Student--")
    for name in rejected_std:
        print(name)

    print("Unique Entries :", len(entered_std))

# Call the function to run the event tracker
run_event_tracker()
```

```python
def reverse_string(s):
    return s[::-1]

# Example
print(reverse_string("hello"))
```

```python
def is_palindrome(s):
    return s == s[::-1]

# Example
print(is_palindrome("madam"))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# Example
print(factorial(5))
```

```python
def count_vowels(s):
    count = 0
    vowels = "aeiouAEIOU"
    for char in s:
        if char in vowels:
            count += 1
    return count

# Example
print(count_vowels("Hello World"))
```

```python
def find_largest(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest

# Example
print(find_largest([3, 7, 2, 9, 5]))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example
print(is_prime(7))
```

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit
        n //= 10
    return total

# Example
print(sum_of_digits(1234))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

# Example
fibonacci(7)
```

```python
def remove_duplicates(lst):
    new_list = []
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    return new_list

# Example
print(remove_duplicates([1, 2, 2, 3, 4, 3]))
```

```python
def string_length(s):
    count = 0
    for _ in s:
        count += 1
    return count

# Example
print(string_length("Python"))
```

```python
#7. Build a Student Record System using nested dictionaries/lists to add students, update marks,
#   compute averages, and find toppers.
students = {}

def add_student(name, marks):
    students[name] = marks

def average(name):
    return sum(students[name]) / len(students[name])

def topper():
    return max(students, key=lambda x: sum(students[x]))

add_student("Aman", [80, 90, 85])
add_student("Ravi", [70, 75, 80])

print("Average Aman:", average("Aman"))
print("Topper:", topper())
```

```python
""" 11 Restaurant Billing System
Display a menu with prices and allow users to order multiple items. Calculate the total bill with
tax. Use loops for ordering, dictionaries for storing menu, and conditionals for bill logic.
 """
 # Restaurant Billing System

# Menu dictionary (item: price)
menu = {
    "pizza": 200,
    "burger": 100,
    "pasta": 150,
    "coffee": 50,
    "sandwich": 80
}

total_bill = 0

print("----- Welcome to Restaurant -----")
print("Menu:")
for item, price in menu.items():
    print(f"{item.capitalize()} : ₹{price}")

# Ordering loop
while True:
    order = input("\nEnter item to order (or 'done' to finish): ").lower()

    if order == "done":
        break

    if order in menu:
        qty = int(input("Enter quantity: "))
        item_total = menu[order] * qty
        total_bill += item_total
        print(f"{order.capitalize()} added! Subtotal: ₹{item_total}")
    else:
        print("Item not available!")

# Billing
print("\n----- Bill Summary -----")
print(f"Total (without tax): ₹{total_bill}")

# Tax calculation (5% GST)
tax = 0.05 * total_bill

# Conditional logic
if total_bill > 500:
    discount = 0.10 * total_bill
    print(f"Discount (10%): -₹{discount}")
else:
    discount = 0

final_bill = total_bill + tax - discount

print(f"Tax (5%): ₹{tax}")
print(f"Final Bill: ₹{final_bill}")

print("Thank you! Visit again 😊")

```
